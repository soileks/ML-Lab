* Нормализация - приведение данных к единичной норме.

* Приведение к отрезку [0;1] - линейное преобразование, приводящее все значения к заданному отрезку.

```python
cols = ['Product Price','Quantity','Total Purchase Amount','Age']
data_numerical = data[cols]
data_numerical.describe()
```

Посмотрим метрики. 

Precision показывает долю правильно предсказанных примеров данного класса из всех примеров, которые модель отнесла к этому классу. Высокая точность означает, что модель почти не делает ложных срабатываний для данного класса.

Recall показывает долю правильно предсказанных примеров данного класса из всех истинных примеров этого класса. Высокая полнота означает, что модель хорошо находит примеры данного класса и не упускает их.

F1-score — это гармоническое среднее между precision (точностью) и recall (полнотой). Эта метрика особенно полезна при несбалансированных данных, так как учитывает как ложные срабатывания (FP), так и пропущенные объекты (FN) для каждого класса. F1-score — идеальная метрика, когда важен баланс между precision и recall, особенно в задачах, где ошибочное предсказание в меньшем классе критично (например, при выявлении мошенничества, диагностики заболеваний, детекции дефектов).

Support показывает, сколько раз каждый класс появляется в истинных метках и служит ориентиром для интерпретации других метрик.

```python
from sklearn.metrics import classification_report

# Выводим precision, recall и F1-score для каждого класса
print(classification_report(y_val, y_val_predict_knn))
```

## Решим задачу кластеризации несколькими алгоритмами:
Kmeans, DBSCAN, OPTICS

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X = data
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

scaler = StandardScaler()
X_scaled_imputed = scaler.fit_transform(X_imputed)
```

```python
inertia = []
clusters = range(2, 50)

for k in clusters:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X_scaled_imputed)
    inertia.append(kmeans.inertia_)

plt.plot(clusters, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

С помощью метода "Локтя" можно увидеть, что начиная с некоторого момента значения меняются не медленно(Можно взять 10 кластеров, но я возьму 30 для эксперимента).

```python
kmeans = KMeans(n_clusters=30, random_state=42, n_init = 'auto')
data['KMeans_Cluster'] = kmeans.fit_predict(X_scaled_imputed)
silhouette_kmeans = silhouette_score(X_scaled_imputed, data['KMeans_Cluster'])
print(f"Silhouette Score for KMeans: {silhouette_kmeans}")
```

В случае, когда мало и истинных значений единиц(отток), и предсказанных, мы потеряем большее количество клиентов, но при этом меньше ложных предсказаний оттока.

Также были рассмотрены и применены три алгоритма кластеризации: KMeans, DBSCAN и OPTICS. Данные были распределены по кластерам с использованием каждого из алгоритмов. Для оценки качества кластеризации использовалась метрика Silhouette Score.

Silhouette score (силуэтный коэффициент) — это метрика, используемая для оценки качества кластеризации. Она измеряет, насколько объекты внутри кластера похожи друг на друга по сравнению с объектами в других кластерах. Силуэтный коэффициент варьируется от -1 до 1. Значение близкое к 1 указывает на хорошую кластеризацию, где объекты внутри кластера очень похожи друг на друга и сильно отличаются от объектов в других кластерах. Значение близкое к 0 указывает на перекрывающиеся кластеры, где объекты могут быть одинаково близки к объектам в других кластерах. Значение близкое к -1 указывает на плохую кластеризацию, где объекты внутри кластера могут быть более похожи на объекты в других кластерах, чем на объекты внутри своего кластера. Лучшее значение получил DBSCAN (OPTICS 0.219, Kmeans 0.16), но это значение 0.22 близко к 0, что говорит о не очень хорошей кластеризации.
